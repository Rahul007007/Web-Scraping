```python
lap = {'Laptop Brand':name, ' Processor Brand ':p_brand, 'Clock Speed' : pclock, 'Cache':cache, 'Number of Cores':num_cores, 'SSD Capacity': ssd_size,'HDD Capacity':hdd_size, 'RAM': r_size, 'Graphic Processor':gpu_desc, 'Dedicated Graphic Memory Capacity':gpu_size, 'Screen Size': screen_size, 'Screen Resolution' : reso, 'Battery Cell': b_cell, 'Weight': wt, 'Cost': cost  }
df = pd.DataFrame.from_dict(lap, orient='index')
df = df.transpose()
df
```

```python
lap = {'Laptop Brand':name, ' Processor Brand ':p_brand, 'Clock Speed' : pclock, 'Cache':cache, 'Number of Cores':num_cores, 'SSD Capacity': ssd_size,'HDD Capacity':hdd_size, 'RAM': r_size, 'Graphic Processor':gpu_desc, 'Dedicated Graphic Memory Capacity':gpu_size, 'Screen Size': screen_size, 'Screen Resolution' : reso, 'Battery Cell': b_cell,'Weight': wt, 'Cost':cost, 'Link': url }
df = pd.DataFrame.from_dict(lap, orient='index')
df = df.transpose()
```
